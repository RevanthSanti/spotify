```python
# Data Cleanig
# Lets copy the data first to avoid dataloss
```

```python
df=data.copy()
```

```python
df.head(5)
```

```python
df.isnull().any()
```

```python
# filling null values of 'Genres' with mode and droping null values from remaining columns
```

```python
df['Genres'] = df['Genres'].fillna(df['Genres'].mode()[0])
```

```python
df.isnull().any()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().any()
```

```python
df
```

```python
# Now lets remove colums which are insignificant to the further analysis
```

```python
df=df.drop(['Track ID','Added By','Time Signature'],axis=1)
```

```python
# correct the date and time format
df
```

```python
df[['Release Date','Duration (ms)','Added At']]
```

```python
# Correct 'Release Date' format 
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce').dt.strftime('%Y-%m-%d') 
# Correct 'Added At' format 
df['Added At'] = pd.to_datetime(df['Added At'], errors='coerce').dt.strftime('%Y-%m-%d %H:%M:%S')
# Convert milliseconds to minutes:seconds 
df['Duration (ms)'] = pd.to_datetime(df['Duration (ms)'], unit='ms').dt.strftime('%M:%S')
```
